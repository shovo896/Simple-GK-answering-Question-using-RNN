tokenization

```python
import re 

def tokenize(text):
       #lowercase 
       text=text.lower()


       # remove quotes 
       text=re.sub(r"[\'\"]", "", text)


       # remove all punctuations 
       text=re.sub(r"[^a-z0-9\s]", "", text)
       text=re.sub(r"\s+", " ", text)

       #remove extra space
       tokens=text.split()


       return tokens 
```

```python
print(tokenize(df['question'][0]))
```
