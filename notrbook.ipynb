tokenization

```python
import re 

def tokenize(text):
       #lowercase 
       text=text.lower()


       # remove quotes 
       text=re.sub(r"[\'\"]", "", text)


       # remove all punctuations 
       text=re.sub(r"[^a-z0-9\s]", "", text)
       text=re.sub(r"\s+", " ", text)

       #remove extra space
       tokens=text.split()


       return tokens 
```

```python
print(tokenize(df['question'][0]))
```

```python
vocab={'UNK': 0}
```

```python
def build_vocab(row):
       print(row['question'],row['answer'])

       tokenized_question = tokenize(row['question'])
       tokenized_answer = tokenize(row['answer'])
       print(tokenized_question,tokenized_answer)
       merged_tokens=tokenized_question+tokenized_answer
       for token in merged_tokens:
              if token not in vocab:
                     vocab[token]=len(vocab)
```

```python
df.apply(build_vocab,axis=1)
```

text to indices 

```python
def text_to_indices(text, word_to_idx=None):
       if word_to_idx is None:
              word_to_idx = vocab

       tokens = tokenize(text) if isinstance(text, str) else text

       indexed_text=[]
       for token in tokens:
              indexed_text.append(word_to_idx.get(token, word_to_idx['UNK']))
       return indexed_text
```

```python
answer_to_idx = {answer: idx for idx, answer in enumerate(df['answer'].astype(str).str.strip().unique())}
idx_to_answer = {idx: answer for answer, idx in answer_to_idx.items()}

class QADataset(Dataset):
       def __init__(self,df,vocab,answer_to_idx):
              self.df = df
              self.vocab = vocab
              self.answer_to_idx = answer_to_idx
       def __len__(self):
              return len(self.df)
       def __getitem__(self,index):
              row = self.df.iloc[index]
              question = text_to_indices(tokenize(row['question']))
              answer = self.answer_to_idx[str(row['answer']).strip()]
              return torch.tensor(question, dtype=torch.long), torch.tensor(answer, dtype=torch.long)
```

```python
dataset=QADataset(df,vocab,answer_to_idx)
```

```python
dataloader=DataLoader(dataset,batch_size=1,shuffle=True)
```

RNN architecture 

```python
import torch.nn as nn 
class simpleRNN(nn.Module):
       def __init__(self,vocab_size,answer_size,embedding_dim=50,hidden_size=64):
              super().__init__()
              self.embedding = nn.Embedding(vocab_size,embedding_dim)
              self.rnn = nn.RNN(embedding_dim,hidden_size,batch_first=True)
              self.fc=nn.Linear(hidden_size,answer_size)
       def forward(self,question):
              embedded=self.embedding(question)
              _,final=self.rnn(embedded)
              return self.fc(final.squeeze(0))
                               
```

```python
model=simpleRNN(vocab_size=len(vocab),answer_size=len(answer_to_idx))
```

```python
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.001)
epochs=50 
```

```python
for epoch in range(epochs):
       total_loss = 0
       for question,answer in  dataloader:
              optimizer.zero_grad()
              output=model(question)
              loss=criterion(output,answer)
              loss.backward()
              optimizer.step()
              total_loss+=loss.item()
       if (epoch+1) % 10 == 0:
              print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(dataloader):.4f}")     
       
       
```

```python
idx_to_word={idx: word for word, idx in vocab.items()}
idx_to_word
```

```python
def predict(question_text,vocab,model,idx_to_answer):
       model.eval()
       with torch.no_grad():
              indices=text_to_indices(question_text, vocab)
              if not indices:
                     return "<UNK>"
              x=torch.tensor(indices, dtype=torch.long).unsqueeze(0)
              pred=torch.argmax(model(x),dim=1).item()
              return idx_to_answer.get(pred,"<UNK>")
```

```python
tests=[
       "What is the capital of France?",
       "Who is the president of the United States?",
       "What is the largest mammal?",]
for test in tests:
       answer=predict(test,vocab,model,idx_to_word)
       print(f"Q: {test}\nA: {answer}\n")
```
